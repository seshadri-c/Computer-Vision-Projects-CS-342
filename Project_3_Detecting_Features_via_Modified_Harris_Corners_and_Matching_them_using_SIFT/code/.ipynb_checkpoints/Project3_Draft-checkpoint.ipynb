```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from utils import plotFigureGray, plotFigure
from student_code import calc_grad_x, calc_grad_y, calc_grad_orientation, calc_grad_mag, harris_corners
```

#### source: https://docs.opencv.org/3.1.0/da/df5/tutorial_py_sift_intro.html 

```python
img1 = cv2.imread('../data/img1.jpg')
img2 = cv2.imread('../data/img2.jpg')



gray1= cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
gray1 = np.float32(gray1)

gray2= cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
gray2 = np.float32(gray2)

```

```python
corners1, Ix1, Iy1 = harris_corners(img1, window_size=3, harris_scoring=0.04, threshold=1e-3, nms_size=5)
corners2, Ix2, Iy2 = harris_corners(img2, window_size=3, harris_scoring=0.04, threshold=1e-3, nms_size=5)

corners1 = cv2.normalize(corners1, corners1, alpha=0, beta=255,
                       norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)
corners2 = cv2.normalize(corners2, corners2, alpha=0, beta=255,
                       norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)
threshold=80


```

```python
def get_keypoints(corners, Ix, Iy, threshold):
    
    rows, cols = np.nonzero(corners > threshold)
    O = calc_grad_orientation(Ix, Iy)
    Mag = calc_grad_mag(Ix, Iy)
    # assign the keypoints
    keypoints = np.zeros((len(rows),))
    keypoints = []
    for i in range(len(rows)):
        r = rows[i]; c = cols[i]
        kp = cv2.KeyPoint(c, r, _size=20, _angle=np.rad2deg(O[r,c]), _octave=0)
        keypoints.append(kp)
        
    return keypoints
```

```python
kp1 = get_keypoints(corners1, Ix1, Iy1, threshold)
kp2 = get_keypoints(corners2, Ix2, Iy2, threshold)
```

```python
cv2.drawKeypoints(img1, kp1, img1,
                          flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plotFigure(img1,12)
plt.show()   
```

```python
cv2.drawKeypoints(img2, kp2, img2,
                          flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plotFigure(img2,12)
plt.show()   
```

```python
# create sift instance
sift = cv2.xfeatures2d.SIFT_create()
# get descriptors
descriptors1 = sift.compute(img1, kp1)[1]
descriptors2 = sift.compute(img2, kp2)[1]



# get matches
bfm = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
matches = bfm.match(descriptors1, descriptors2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = np.array([])

img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:10], img3, flags=2)

```

```python
plotFigure(img3,24)
plt.show() 
```
